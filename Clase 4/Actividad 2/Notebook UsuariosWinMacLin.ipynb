```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#  Cargar datos desde el CSV 
data = pd.read_csv("usuarios_win_mac_lin.csv")

# Revisar las primeras filas
print(data.head())
```

```python
# Separar características y etiqueta
X = data[['duracion', 'paginas', 'acciones', 'valor']]
y = data['clase']
```

```python
# Estandarizar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Dividir en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

```python
# Entrenar modelo de Regresión Logística
model = LogisticRegression(multi_class='multinomial', solver='lbfgs')
model.fit(X_train, y_train)
```

```python
# Predicciones con datos de prueba
y_pred = model.predict(X_test)
print("Precisión del modelo:", accuracy_score(y_test, y_pred))
```

```python
# Alterar valores de un usuario y predecir
usuario_modificado = X_test[0].copy()
usuario_modificado[0] *= 2  # duplicar duración
usuario_modificado[1] += 1  # agregar 1 página vista
```

```python
pred_modificado = model.predict([usuario_modificado])
print("Predicción del usuario modificado:", pred_modificado)
```

```python
ejemplo = [[7, 2, 4, 8]]  # duracion, paginas, acciones, valor
prediccion = model.predict(ejemplo)
print(prediccion)
```

```python
ejemplo_modificado = [[12, 5, 2, 2]]
prediccion_modificada = model.predict(ejemplo_modificado)
print(prediccion_modificada)
```

El objetivo de esta actividad fue clasificar a los usuarios de un sitio web según su sistema operativo (Windows, Macintosh o Linux) usando un modelo de Regresión Logística basado en 4 características extraídas de Google Analytics:

Duración de la visita (segundos)

Cantidad de páginas vistas

Cantidad de acciones del usuario

Suma del valor de las acciones

La salida se codificó de la siguiente manera:

0 → Windows

1 → Macintosh

2 → Linux

Se utilizó la notebook Regresion_logistica.ipynb, siguiendo los pasos:

Carga y exploración del dataset usuarios_win_mac_lin.csv.

Separación de las características (X) y la etiqueta (y).

División del dataset en conjunto de entrenamiento y prueba.

Entrenamiento del modelo de Regresión Logística.

Evaluación de la precisión y análisis de predicciones.

Experimentación alterando los valores de las variables para observar el comportamiento de la clasificación.

Precisión del modelo: El modelo mostró una precisión adecuada en el conjunto de prueba, indicando que las 4 características seleccionadas son relevantes para diferenciar entre los sistemas operativos.

Comportamiento ante cambios en las variables:

Aumentar la duración de la visita o la cantidad de páginas vistas tendió a aumentar la probabilidad de que el usuario fuera clasificado como Linux, dado que los usuarios de Linux del dataset tienden a sesiones más largas y exploratorias.

Incrementos en acciones y valor de acciones favorecieron la predicción de Macintosh, reflejando un comportamiento más interactivo de estos usuarios.

Valores extremos o inconsistentes pueden provocar que el modelo clasifique erróneamente, especialmente si los datos no son representativos.

Interpretación de coeficientes: Los coeficientes de la regresión logística muestran qué variables tienen mayor influencia en cada clase. Esto permite entender qué factores diferencian más a los usuarios según su sistema operativo.

En conclusión, la Regresión Logística es efectiva para clasificar usuarios según su sistema operativo usando solo métricas de interacción del sitio web.

Las variables de comportamiento (duración, páginas vistas, acciones y valor de acciones) son predictivas y permiten identificar patrones característicos de cada sistema operativo.

La experimentación con valores alterados evidencia la sensibilidad del modelo a cambios en los datos, lo que es útil para simular escenarios y comprender la robustez del clasificador.

```python
# Ejercicio 7: Multiplica dos matrices 3x3 de enteros aleatorios entre 1 y 10.
import numpy as np

# Crear matrices 3x3 con enteros aleatorios entre 1 y 10
A = np.random.randint(1, 11, (3, 3))
B = np.random.randint(1, 11, (3, 3))

# Multiplicación matricial
resultado = np.dot(A, B)   # también podés usar A @ B

print("Matriz A:\n", A)
print("Matriz B:\n", B)
print("Resultado:\n", resultado)
```

```python
# Ejercicio 8: Crea un array de 15 valores aleatorios entre 1 y 100 e imprime su máximo, mínimo y promedio.
import numpy as np

# Crear array
array = np.random.randint(1, 101, 15)

# Calcular estadísticas
maximo = np.max(array)
minimo = np.min(array)
promedio = np.mean(array)

print("Array:", array)
print("Máximo:", maximo)
print("Mínimo:", minimo)
print("Promedio:", promedio)
```

```python
# Ejercicio 9: Genera una matriz 5x5 con valores del 1 al 25 y extrae su diagonal.
import numpy as np

# Crear matriz 5x5 con valores del 1 al 25
matriz = np.arange(1, 26).reshape(5, 5)

# Extraer diagonal
diagonal = np.diag(matriz)

print("Matriz:\n", matriz)
print("Diagonal:", diagonal)
```

```python
# Ejercicio 10 : Dada una matriz 5x5 de enteros aleatorios entre 1 y 20, calcula la suma de cada fila y columna
import numpy as np

# Crear matriz 5x5 con enteros aleatorios entre 1 y 20
matriz = np.random.randint(1, 21, (5, 5))

# Suma por filas (axis=1)
suma_filas = np.sum(matriz, axis=1)

# Suma por columnas (axis=0)
suma_columnas = np.sum(matriz, axis=0)

print("Matriz:\n", matriz)
print("Suma de filas:", suma_filas)
print("Suma de columnas:", suma_columnas)
```

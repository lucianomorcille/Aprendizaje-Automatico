```python
import pandas as pd

# Cargar el archivo
df = pd.read_csv("Automobile.csv")

# Ver primeras filas
print(df.head())
```

```python
# Grafico mpg vs weight
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(df["weight"], df["mpg"])
plt.xlabel("Weight")
plt.ylabel("MPG")
plt.title("Relación entre peso y consumo")
plt.show()
```

```python
# Histograma mpg
plt.figure()
plt.hist(df["mpg"], bins=10, color= "orange")
plt.xlabel("MPG")
plt.ylabel("Frecuencia")
plt.title("Distribución del consumo de combustible")
plt.show()
```

```python
df.boxplot(column="cylinders", by="origin")
plt.xlabel("origen")
plt.ylabel("cilindradas")
plt.title("Cilindradas por origen")
plt.show()
```

```python
grupo = df.groupby(["model_year", "origin"]).size().unstack()

grupo.plot(kind="bar")
plt.xlabel("Año del modelo")
plt.ylabel("Cantidad de autos")
plt.title("Distribución por año y origen")
plt.show()
```
